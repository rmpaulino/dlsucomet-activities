```python
import networkx as nx
import matplotlib.pyplot as plt 
```

```python
G = nx.Graph()
node_list = [
    ("node 1", {"color": "red"}),
    "node 2",
    "node 3"
]
G.add_nodes_from(node_list)
print(G.nodes)
```

```python
G.add_edges_from([
    ("node 1", "node 3", {"weight" : 3}),
    ("node 2", "node 7", {"weight" : 5}),
    ("node 1", "node 2", {"weight" : 7}),
    ("node 2", "node 6"),
    ("node 4", "node 5"),
    ("node 4", "node 7"),
    ("node 7", "node 1")
])

weights = [1 if G[u][v] == {} else G[u][v] ['weight'] for u,v in G.edges()]
# print (G.edges())
# print (weights)

print ("Nodes: ", G.nodes, "\nEdges: ", G.edges)
print (G)

nx.draw(G, width=weights, with_labels = True)
```
